```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# creates a pandas dataframe from the csv file
data = pd.read_csv('Vital_Statistics_Deaths_by_Region_and_Age-Group_by_Selected_Cause_of_Death___Beginning_2003.csv')

# displays the data
data
```

```python
# displays the series of unique elements from the column 'Selected Cause of Death'
data['Selected Cause of Death'].unique()
```

```python
# remove rows that contain 'Total'
data = data[data['Selected Cause of Death'] != 'Total']

# creates a series of unique elements from the column 'Selected Cause of Death' 
# and converts it into a list and assigns it to cause
cause = data['Selected Cause of Death'].unique().tolist()
cause.sort()

# displays the list
cause
```

```python
# a pandas series of the unqiue elements in the colum 'Race or Ethnicity' is converted into a python list and assigned to re
re = data['Race or Ethnicity'].unique().tolist()

# sorts the list
re.sort()

# displays the list
re
```

```python
print('pandas array: {}'.format(data['Year'].unique()))
# creates a list of all the unique years from the column 'Year'
years = data['Year'].unique().tolist()

# reverses the list
years.reverse()
print(years)
```

```python
# displays the year and the total deaths from that year
data.groupby('Year')['Deaths'].sum()
```

```python
year_deaths = data.groupby('Year')['Deaths'].sum()

# creates a figure with a custom figure size
plt.figure(figsize=(10, 5))

# creates bar chart with the Year as the x-axis and the number of deaths as the y-axis
plt.bar(years, year_deaths)
plt.title("Fig. 1: Total deaths from 2003 to 2016", fontweight='bold')
plt.xticks(years)
plt.xlabel("Year", fontweight='bold')
plt.ylabel("Number of deaths", fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(10,4))

year_deaths = year_deaths.to_list()
year_deaths_percent_change = []

for x in range(len(year_deaths)):
    year_deaths_percent_change.append((year_deaths[x] - year_deaths[0]) / year_deaths[0] * 100)
    
plt.plot(years, year_deaths_percent_change)
plt.title('Fig. 2: Percent of Change of Total Deaths From 2003 to 2016', fontweight='bold')
plt.xticks(years)
plt.xlabel('Year', fontweight='bold')
plt.ylabel('Percent Change', fontweight='bold')
plt.show()
```

```python
#sums the number of deaths and groups them by race or ethnicity
data.groupby('Race or Ethnicity')['Deaths'].sum()
```

```python
plt.figure(figsize=(12, 5))

re_deaths = data.groupby('Race or Ethnicity')['Deaths'].sum()

plt.bar(re, re_deaths)
plt.title('Fig. 3: Number of Deaths by Race or Ethnicity', fontweight='bold')
plt.xlabel('Race or Ethnicity', fontweight='bold')
plt.ylabel('Number of deaths', fontweight='bold')
plt.show()
```

```python
data.groupby('Selected Cause of Death')['Deaths'].sum()
```

```python
cause_deaths = data.groupby('Selected Cause of Death')['Deaths'].sum()

plt.figure(figsize=(20, 5))
plt.bar(cause, cause_deaths)
plt.title('Fig. 4: Number of Deaths by Cause', fontweight='bold', fontsize=16)
plt.xlabel('Selected Cause of Death', fontweight='bold', fontsize=14)
plt.xticks(fontsize=10)
plt.ylabel('Number of deaths', fontweight='bold', fontsize=14)
plt.show()
```

```python
data.groupby(['Selected Cause of Death', 'Race or Ethnicity'])['Deaths'].sum()
```

```python
cause_re_deaths = data.groupby(['Selected Cause of Death', 'Race or Ethnicity'])['Deaths'].sum()
cause_re_bars = []

for i in range(len(re)):
    cause_re_bars.append([])
    for j in range(i, len(cause_re_deaths), len(re)):
        cause_re_bars[i].append(cause_re_deaths[j])

xticks = np.arange(len(cause))
bar_width = 1/(len(cause_re_bars) + 1)
bar_pos = []

for i in range(len(cause_re_bars)):
    bar_pos.append([])
    if(i == 0):
        bar_pos[i] = list(xticks)
    else:
        bar_pos[i] = [x + bar_width for x in bar_pos[i - 1]]
        
plt.figure(figsize=(20, 5))

for i in range(len(cause_re_bars)):
    plt.bar(bar_pos[i], cause_re_bars[i], width=bar_width, label=re[i])

plt.title('Fig. 5: Deaths by Cause Grouped by Race or Ethnicity', fontweight='bold', fontsize=16)
plt.ylabel('Number of Deaths', fontweight='bold', fontsize=14)
plt.yticks(fontsize=12)
plt.xlabel('Race or Ethnicity', fontweight='bold', fontsize=14)
plt.xticks(bar_pos[int(len(bar_pos)/2)], cause, fontsize=10)
plt.legend()
plt.show()
```

```python
#groups by year and then race or ethnicity and sums the number of deaths
data.groupby(['Year', 'Race or Ethnicity'])['Deaths'].sum()
```

```python
year_re_deaths = data.groupby(['Year', 'Race or Ethnicity'])['Deaths'].sum()

year_re_bars = []
for i in range(len(re)):
    year_re_bars.append([])
    
i = 0
j = i
while i < len(year_re_deaths):
    if i != 35:
        year_re_bars[j].append(year_re_deaths[i])
        year_re_bars[j + 1].append(year_re_deaths[i + 1])
        year_re_bars[j + 2].append(year_re_deaths[i + 2])
        year_re_bars[j + 3].append(year_re_deaths[i + 3])
        year_re_bars[j + 4].append(year_re_deaths[i + 4])
        i += 5
    else:
        year_re_bars[j].append(year_re_deaths[i])
        year_re_bars[j + 1].append(year_re_deaths[i + 1])
        year_re_bars[j + 2].append(year_re_deaths[i + 2])
        year_re_bars[j + 4].append(year_re_deaths[i + 3])
        i += 4
        
year_re_bars[3].insert(7, 1)
        
xticks = np.arange(len(years))
bar_width = 1/(len(year_re_bars) + 1)
bar_pos = []

for i in range(len(year_re_bars)):
    bar_pos.append([])
    if(i == 0):
        bar_pos[i] = list(xticks)
    else:
        bar_pos[i] = [x + bar_width for x in bar_pos[i - 1]]

plt.figure(figsize=(16, 6))

for i in range(len(bar_pos)):
    plt.bar(bar_pos[i], year_re_bars[i], width=bar_width, label=re[i])

plt.title('Fig. 6: Number of Deaths in NY by Race or Ethnicity from 2003 to 2016', fontweight='bold', fontsize=16)
plt.ylabel('Number of Deaths', fontweight='bold', fontsize=14)
plt.yticks(fontsize=12)
plt.xlabel('Year', fontweight='bold', fontsize=14)
plt.xticks(bar_pos[int(len(bar_pos)/2)], years, fontsize=12)
plt.legend()
plt.show()
```

```python
year_tail = years[-2:]
year_re_bars_tail = []
start = len(year_re_deaths) - 10
for i in range(len(re)):
    year_re_bars_tail.append([])
    for j in range(start + i, len(year_re_deaths), len(re)):
        year_re_bars_tail[i].append(year_re_deaths[j])

xticks = np.arange(len(year_tail))
bar_width = 1/(len(year_re_bars_tail) + 1)
bar_pos = []

for i in range(len(year_re_bars_tail)):
    bar_pos.append([])
    if(i == 0):
        bar_pos[i] = list(xticks)
    else:
        bar_pos[i] = [x + bar_width for x in bar_pos[i - 1]]

for i in range(len(bar_pos)):
    plt.bar(bar_pos[i], year_re_bars_tail[i], width=bar_width, label=re[i])

plt.title('Fig. 7: Number of Deaths in NY by Race or Ethnicity from 2015 to 2016', fontweight='bold')
plt.ylabel('Number of Deaths', fontweight='bold')
plt.xlabel('Year', fontweight='bold')
plt.xticks(bar_pos[int(len(bar_pos)/2)], year_tail)
plt.legend(loc='upper center', bbox_to_anchor=(1.225, 0.75))
plt.show()
```

```python
year_re_deaths_percent_change = []

for i in range(len(year_re_bars)):
    year_re_deaths_percent_change.append([])
    for j in range(len(year_re_bars[i])):
        year_re_deaths_percent_change[i].append((year_re_bars[i][j] - year_re_bars[i][0]) / year_re_bars[i][0] * 100)

plt.figure(figsize=(8,4))
        
for i in range(len(re)):
    plt.plot(years, year_re_deaths_percent_change[i], label=re[i])
    
plt.title('Fig. 8: Percent Change of Deaths by Race or Ethnicity From 2003 to 2016', fontweight='bold')
plt.xticks(years)
plt.xlabel('Year', fontweight='bold')
plt.ylabel('Percent of Change', fontweight='bold')
plt.legend()
plt.show()
```

```python
year_re_deaths_percent_change = []

for i in range(len(year_re_bars)):
    year_re_deaths_percent_change.append([])
    for j in range(len(year_re_bars[i]) - 2):
        year_re_deaths_percent_change[i].append((year_re_bars[i][j] - year_re_bars[i][0]) / year_re_bars[i][0] * 100)

years_head = years[:12]

plt.figure(figsize=(8,4))
        
for i in range(len(re)):
    plt.plot(years_head, year_re_deaths_percent_change[i], label=re[i])
    
plt.title('Fig 9: Percent Change of Deaths of Races or Ethnicities From 2003 to 2014', fontweight='bold')
plt.xticks(years_head)
plt.xlabel('Year', fontweight='bold')
plt.ylabel('Percent of Change', fontweight='bold')
plt.legend()
plt.show()
```
